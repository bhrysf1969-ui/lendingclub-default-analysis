🎯 Insight
Our analysis of Lending Club loan data (2007–2015) shows that loan grade is strongly correlated with the likelihood of default (Charged Off).
Lower-grade loans (Grades E, F, and G) have significantly higher default rates compared to higher grades (A and B).
While these lower-grade loans carry higher interest rates, the increased revenue from interest does not fully offset the higher loss rates, making them risk-heavy segments in the portfolio.
    
		Example insight:
		Grade E default rate ≈ 26%
		Grade F default rate ≈ 34%
		Grade G default rate ≈ 37%
This pattern indicates that credit risk escalates rapidly in lower grades, suggesting the current underwriting and pricing for these segments may not be sustainable.


    Recommendation:

Tighten underwriting criteria or adjust pricing for lower loan grades (D–F) to balance risk and profitability.

	Implementation Options:
    
1. Underwriting Adjustment:
    ○ Increase the minimum FICO score or reduce maximum debt-to-income ratio for applicants likely to fall into D–F grades.
    ○ This reduces the inflow of high-risk borrowers.
2. Pricing Adjustment:
	○ Slightly increase interest rates for lower grades to better cover expected losses.
	○ Alternatively, limit the volume of loans issued in these grades.
3. Monitoring:
	○ Track charge-off rates monthly using the loan_default_summary table (from Part 1) to detect early warning trends.


	Business Impact:
• Reduced loan losses and credit provisioning costs.
• Improved net yield by aligning pricing and risk exposure.
• Stronger credit risk governance and confidence from investors.
